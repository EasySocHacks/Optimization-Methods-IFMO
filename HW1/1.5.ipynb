```python

import warnings

import numpy as np

from HW1.gradient_descent import gradient_descent
from HW1.scheduler import ExponentialScheduler, StepScheduler
from HW1.visualize import draw_3D
```

# Функция 1

$f1(x) =$ ${x^2 + y^2} \over 200$


```python
def f1(coord):
    return (coord[0] ** 2 + coord[1] ** 2) / 200
```

```python
draw_3D(f1)
```

```python
with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=RuntimeWarning)
    (xs, y), meta = gradient_descent(f1, 2)
    print(xs, y)
    print(meta)

    draw_3D(f1, meta["points"])
```

```python
with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=RuntimeWarning)
    (xs, y), meta = gradient_descent(f1, 2, coord=np.array([-90, -90]))
    print(xs, y)
    print(meta)

    draw_3D(f1, meta["points"])
```

```python
with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=RuntimeWarning)
    (xs, y), meta = gradient_descent(f1, 2, coord=np.array([-90, -90]), lr=0.9)
    print(xs, y)
    print(meta)

    draw_3D(f1, meta["points"])
```

```python
with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=RuntimeWarning)
    (xs, y), meta = gradient_descent(f1, 2, coord=np.array([5, 5]))
    print(xs, y)
    print(meta)

    draw_3D(f1, meta["points"])
```

```python
with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=RuntimeWarning)
    (xs, y), meta = gradient_descent(f1, 2, lr=0.9, coord=np.array([5, 5]))
    print(xs, y)
    print(meta)

    draw_3D(f1, meta["points"])
```

```python
with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=RuntimeWarning)
    (xs, y), meta = gradient_descent(f1, 2, coord=np.array([90, 90]), lr=0.9, scheduler=ExponentialScheduler(50, -0.05))
    print(xs, y)
    print(meta)

    draw_3D(f1, meta["points"])
```

```python
with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=RuntimeWarning)
    (xs, y), meta = gradient_descent(f1, 2, coord=np.array([90, 90]), lr=0.9, scheduler=StepScheduler(50, -0.05))
    print(xs, y)
    print(meta)

    draw_3D(f1, meta["points"])
```

## Итого
На даннй функции можно выделить следующие особенности:
> Лучше выбирать $lr$ побольше (например, $0.9$)

> Скорость сходимости не так сильно зависит от выбора изначальной точки

> $Scheduler$'ы хорошо ускоряют время сходимсоти

# Функция 2

$f2(x) = |{(x^2 - y) * cos({x \over 50}) \over 500}|$

```python
def f2(coord):
    return np.abs(((coord[0] ** 2 - coord[1]) * np.cos(coord[0] / 50)) / 500)
```

```python
draw_3D(f2)
```

```python
with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=RuntimeWarning)
    (xs, y), meta = gradient_descent(f2, 2)
    print(xs, y)
    print(meta)

    draw_3D(f2, meta["points"])
```

```python
with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=RuntimeWarning)
    (xs, y), meta = gradient_descent(f2, 2, np.array([-99, 0]))
    print(xs, y)
    print(meta)

    draw_3D(f2, meta["points"])
```

```python
with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=RuntimeWarning)
    (xs, y), meta = gradient_descent(f2, 2, lr=0.9, coord=np.array([-99, 0]))
    print(xs, y)
    print(meta)

    draw_3D(f2, meta["points"])
```

```python
with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=RuntimeWarning)
    (xs, y), meta = gradient_descent(f2, 2, lr=0.9, coord=np.array([-50, 0]))
    print(xs, y)
    print(meta)

    draw_3D(f2, meta["points"])
```

```python
with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=RuntimeWarning)
    (xs, y), meta = gradient_descent(f2, 2, check_batch=10, lr=0.9, coord=np.array([-100, 0]), scheduler=ExponentialScheduler(100, -0.5))
    print(xs, y)
    print(meta)

    draw_3D(f2, meta["points"])
```

```python
with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=RuntimeWarning)
    (xs, y), meta = gradient_descent(f2, 2, check_batch=10, lr=0.9, coord=np.array([-100, 0]), scheduler=StepScheduler(100, 0.05))
    print(xs, y)
    print(meta)

    draw_3D(f2, meta["points"])
```

## Итого
На даннй функции можно выделить следующие особенности:
> Лучше выбирать $lr$ побольше (например, $0.9$)

> Скорость сходимости сильно зависит от изначальной точки.
> При достижении $\pm$ локального минимума, мы попадаем на плато, и при больших $lr$ продолжаем движение.

> $Scheduler$'ы ускоряют время сходимсоти, но нужно подбирать под конкретную ситуацию.
