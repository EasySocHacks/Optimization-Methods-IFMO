```python
import warnings

import numpy as np

from HW1.gradient_descent import gradient_descent
from HW1.visualize import draw_2D, draw_3D
from scheduler import *
```

# Example 1

> $dim = 1$.
>
> $f(x) = {x^4 \over 1e7}$

```python
def f(coord):
    return (coord ** 4) / 1e7
```

```python
draw_2D(f)
```

```python
with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=RuntimeWarning)
    (x, y), meta = gradient_descent(f, 1, lr=0.9, scheduler=ExponentialScheduler(10, -0.2))
    print(x, y)
    print(meta)

    draw_2D(f, meta["points"])
```

```python
with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=RuntimeWarning)
    (x, y), meta = gradient_descent(f, 1, lr=0.9, scheduler=StepScheduler(50, 0.04))
    print(x, y)
    print(meta)

    draw_2D(f, meta["points"])
```

# Example 2

> $dim = 2$.
>
> $f(x, y) = sin(x / 20) * cos(e + y / 20)$

```python
def f(coord):
    return np.sin(coord[0] / 20) * np.cos(np.e + coord[1] / 20)
```

```python
draw_3D(f)
```

```python
with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=RuntimeWarning)
    (xs, y), meta = gradient_descent(f, 2, lr=0.9, scheduler=ExponentialScheduler(100, -0.5))
    print(xs, y)
    print(meta)

    draw_3D(f, meta["points"])
```

```python
with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=RuntimeWarning)
    (xs, y), meta = gradient_descent(f, 2, lr=0.9, scheduler=StepScheduler(25, 0.004))
    print(xs, y)
    print(meta)

    draw_3D(f, meta["points"])
```
