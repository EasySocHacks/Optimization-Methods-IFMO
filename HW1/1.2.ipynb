```python
import warnings

from grad import *
from scheduler import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
dim = 1


def f(coord):
    if coord >= 0:
        return coord - np.sin(coord)
    else:
        return -coord
```

```python
xs = np.linspace(-100, 100, 1000)
ys = list(map(f, xs))

plt.plot(xs, ys)
plt.show()
```

```python
with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=RuntimeWarning)
    (x, y), meta = gradient_descent(f, 1, lr=0.9, scheduler=ExponentialScheduler(10, -0.2))
    print(x, y)
    print(meta)


    plt.plot(meta["points"][:, 0], meta["points"][:, 1], 'r.')
    plt.plot(xs, ys)
    plt.show()
```

```python
with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=RuntimeWarning)
    (x, y), meta = gradient_descent(f, 1, lr=0.9, scheduler=StepScheduler(50, 0.04))
    print(x, y)
    print(meta)


    plt.plot(meta["points"][:, 0], meta["points"][:, 1], 'r.')
    plt.plot(xs, ys)
    plt.show()
```
